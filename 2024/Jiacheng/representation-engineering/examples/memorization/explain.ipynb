```python
import torch
from transformers import AutoModelForQuestionAnswering, AutoTokenizer
import matplotlib.pyplot as plt
import seaborn as sns

class CustomQAViz:
    def __init__(self, model_name):
        self.tokenizer = AutoTokenizer.from_pretrained(model_name)
        self.model = AutoModelForQuestionAnswering.from_pretrained(model_name, output_attentions=True)

    def get_attention_weights(self, question, context):
        inputs = self.tokenizer(question, context, return_tensors='pt')
        outputs = self.model(**inputs)

        # 获取最后一层的注意力权重
        attention = outputs.attentions[-1]
        return attention, inputs

    def visualize_attention(self, attention, inputs):
        # 选择第一个头的注意力权重
        attention = attention[0, 0].detach().numpy()

        # 获取token
        tokens = self.tokenizer.convert_ids_to_tokens(inputs['input_ids'][0])

        # 绘制热力图
        plt.figure(figsize=(10, 10))
        sns.heatmap(attention, xticklabels=tokens, yticklabels=tokens, cmap='viridis')
        plt.xlabel('Tokens')
        plt.ylabel('Tokens')
        plt.title('Attention Weights')
        plt.show()

# 使用示例
model_name = "meta-llama/Meta-Llama-3-8B-Instruct"
qa_viz = CustomQAViz(model_name)

question = "What is the capital of France?"
context = "France is a country in Europe. The capital of France is Paris."

attention, inputs = qa_viz.get_attention_weights(question, context)
qa_viz.visualize_attention(attention, inputs)
```
